```python
import pandas as pd
import seaborn as sns
```

```python
df=pd.read_csv("StudentsPerformance 2.08.18 AM.csv")
```

```python
df.info()
```

```python
df['math score'].fillna(df['math score'].mean(),inplace=True,axis=0)
df['reading score'].fillna(df['reading score'].mean(),inplace=True, axis=0)
df['writing score'].fillna(df['writing score'].mean(),inplace=True, axis=0)
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df.boxplot()
```

```python
df['math score']=df['math score'].astype(int)
df['reading score']=df['reading score'].astype(int)
df['writing score']=df['writing score'].astype(int)
```

```python
df['Zscore']=(df['math score'] - df['math score'].mean())/df['math score'].std()  #zscore-normalization
```

```python
df.head()
```

```python
 df_no_outlier=df[(df['Zscore']<3) & (df['Zscore']> -3)] 
```

```python

df_no_outlier
```

```python
min_t,max_t=df_no_outlier['reading score'].quantile([0.001,0.999])
```

```python
df_min_max=df_no_outlier[(df_no_outlier['reading score']>min_t) & (df_no_outlier['reading score'] < max_t)]
```

```python
df_min_max
```

```python
q1=df.quantile(q=0.25,  numeric_only=True)
q1
q3=df.quantile(q=0.75,numeric_only=True)
q3
```

```python
IQR=q3-q1
IQR
```

```python
LL=q1-(IQR*1.5)
UL=q3+(IQR*1.5)
UL
```

```python
df.info()
```

```python
LD=df[['math score','reading score','writing score']]< LL
LD
```

```python
UD=df[['math score','reading score','writing score']] > UL
```

```python
df[(LD | UD)]
```

```python
out=df[~(LD | UD).any(axis=1)]
```

```python
out
```
